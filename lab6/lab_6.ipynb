У попередньому завданні вам потрібно було створити:

- карту, котра відображає рівень приросту / скорочення населення за регіонами станом на останній доступний рік;

- лінійний графік, котрий відображає рівень приросту / скорочення населення за регіонами за всіма роками (одна лінія на графіку - один регіон).

Для цього завдання вам потрібно налагодити інтеракцію між картою та графіком.

При наведенні на якийсь регіон на карті він має виділятися на графіку за допомогою товщини лінії та її кольору.

При наведенні на якусь лінію на графіку відповідний їй регіон має виділятися на карті за допомогою прозорості.

```python
import altair as alt
import pandas as pd
import altair_viewer

alt.renderers.enable('altair_viewer')
```

```python
data = pd.read_csv('population_trends.csv')
data.head()
```

```python
import geopandas as gpd
ukraine = gpd.read_file('ukraine.json')
ukraine.head()
```

```python
diff = list(set(ukraine['NAME_1']).symmetric_difference(set(data['region'])))
data_2019 = data[data['year']==2019]
data_2019 = data_2019[data_2019['region']!=diff[0]]


cols = [i for i in ukraine.columns]
cols[3] = 'region'
ukraine.columns = cols
ukraine.head()
```

```python
merged = ukraine.merge(data_2019, how='outer')
merged.head()
```

```python
data_2019= data_2019.reset_index().drop(columns=['index'])
```

```python
import numpy as np
average = 0
lst = []
for i,r in merged.iterrows():
    if np.isnan(r['rate']):
        r['rate'] = average
        lst.append("None")
        merged.iloc[i,:] = r
    else:
        lst.append(str(r['rate']))
merged['str_rate'] = lst
```

```python
merged
```

```python
data['Year'] = data['year'].astype('str')
data.head()
```

```python
merged["year"] = merged["year"].astype("str")
```

```python
select_region_map = alt.selection_single(on = 'mouseover', fields = ['region'], nearest = False, empty = 'all')
select_region_lines = alt.selection_single(on = 'mouseover', fields = ['region'], nearest = False, empty = 'all')

country_map = alt.Chart(merged, title='Population trends in Ukraine in 2019').project(type = 'equalEarth'
    ).mark_geoshape().encode(
    color = alt.Color('rate:Q', title='Rate of population change'),
    tooltip = [
        alt.Tooltip('region:O', title='Region name'),
        alt.Tooltip('str_rate:O', title='Population increase rate in 2019')
    ],
    opacity = alt.condition((select_region_map) | (select_region_lines), alt.value(1), alt.value(0.6))
).add_selection(
    select_region_map
).properties(width = 700, height = 400)


side_line_plot = alt.Chart(data, title='Population trends in Ukraine through all the years').mark_line().encode(
    x = alt.X(field='Year', 
              type='temporal'
             ),
    y = alt.Y('rate:Q'),
#     color = alt.Color('region:O', scale = alt.Scale(type = 'quantize'),title='Region name'),
    detail = alt.Detail('region:N'),
    tooltip = [
        alt.Tooltip('region:O', title='Region name')
    ],
    size = alt.condition(
    (select_region_map) | (select_region_lines),
    alt.value(4),
    alt.value(0.2)),
    color = alt.condition((select_region_lines) | (select_region_map), 
                          alt.Color('region:O', scale = alt.Scale(type = 'quantize'),title='Region name'), 
                          alt.value("darkgrey"))
).add_selection(select_region_lines
).properties(width = 600, height = 400)


alt.hconcat(country_map, 
            side_line_plot
           ).configure_view(
    strokeWidth=0
).configure_title(fontSize=17)
```
