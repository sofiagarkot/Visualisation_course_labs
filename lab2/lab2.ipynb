```python
year_index_correspondance = {}
for i, r in data.iterrows():
    y = r['Час/Дата'][-4:]
    if y not in year_index_correspondance:
        year_index_correspondance[y] = [i]
    else:
        year_index_correspondance[y].append(i)
```

Циклічне прпдставлення недостатньо підходить для тижня, так як тиждень зазвичай сприймається лінійно а не циклічно. У хімапі не було потреби, так як розмірність даних 2.

```python
weekday_consumption_correspondance = {}
stdevs = [None for i in range(7)]
for i, r in data.iterrows():
    hour = int(r['Час/Дата'][-13:-11])
    month = int(r['Час/Дата'][-7:-5])
    hour = int(r['Час/Дата'][-13:-11])
    day = int(r['Час/Дата'][-10:-8])
    month = int(r['Час/Дата'][-7:-5])
    y = int(r['Час/Дата'][-4:])
    
    
    weekday = date(y, month, day).weekday()
    
    if weekday not in weekday_consumption_correspondance:
        weekday_consumption_correspondance[weekday]= [r['CONSUMPTION']]
    else:
        weekday_consumption_correspondance[weekday].append(r['CONSUMPTION'])
    
for w in weekday_consumption_correspondance:
    stdevs[w] = np.std(np.asarray(weekday_consumption_correspondance[w]))
    weekday_consumption_correspondance[w] = np.average(np.asarray(weekday_consumption_correspondance[w]))
weekday_consumption_correspondance
```
