```python
import altair as alt
import pandas as pd
import altair_viewer

alt.renderers.enable('altair_viewer')
```

```python
data = pd.read_csv('population_trends.csv')
data.head()
```

```python
data.shape
```

```python
set(data['region'])
```

```python
import geopandas as gpd
ukraine = gpd.read_file('ukraine.json')
ukraine.head()
```

```python
ukraine.shape
```

```python
diff = list(set(ukraine['NAME_1']).symmetric_difference(set(data['region'])))
```

```python
new_df = data[data['year']==2019]
new_df = new_df[new_df['region']!=diff[0]]
```

```python
cols = [i for i in ukraine.columns]
cols[3] = 'region'
ukraine.columns = cols
ukraine.head()
```

```python
merged = ukraine.merge(new_df)
merged.head()
```

```python
import numpy as np
average = 0
lst = []
for i,r in merged.iterrows():
    if np.isnan(r['rate']):
        r['rate'] = average
        lst.append("None")
        merged.iloc[i,:] = r
    else:
        lst.append(str(r['rate']))
merged['str_rate'] = lst
```

```python
merged.head()
```

```python
data2019 = data[data['year'] == 2019]
data2019
```

```python
data.head()
```

```python
cols_new_df = list(set(merged['region']))
cols_new_df.append('year')
cols_new_df
```

```python
year_region_rate = {}
for i, r in data.iterrows():
    if r['region'] != 'Ukraine':
        region = r['region']
        year = r['year']
        rate = r['rate']
        if year not in year_region_rate:
            year_region_rate[year] = {}
        
        if not np.isnan(rate):
            year_region_rate[year][region] = rate
        else:
            year_region_rate[year][region] = average
        
        
```

```python
final = [cols_new_df]
for y in year_region_rate:
    one_year_data = []
    for i in cols_new_df[:-1]:
#         print("i", i, 'y', y, 'r', r)
        one_year_data.append(year_region_rate[y][i])
        
    one_year_data.append(y)
    final.append(one_year_data)

d2 = pd.DataFrame(data=final[1:], columns=final[0])
d2.head()
```

```python
data["Year"] = data["year"].astype("str")
```

```python
country_map = alt.Chart(merged, title='Population trends in Ukraine in 2019').project(type = 'equalEarth').mark_geoshape().encode(
    color = alt.Color('rate:Q', title='Rate of population change', scale=alt.Scale(scheme='reds', reverse=True), legend=alt.Legend(title='Rate')),
    tooltip = [
        alt.Tooltip('region:O', title='Region name'),
        alt.Tooltip('str_rate:O', title='Population increase rate in 2019')
    ]
).properties(width = 700, height = 400)


```

```python
side_line_plot = alt.Chart(data, title='Population trends in Ukraine through all the years').mark_line().encode(
    x = alt.X(field='Year', 
              type='temporal'
             ),
    y = alt.Y('rate:Q'),
    color = alt.Color('region:O', scale = alt.Scale(type = 'quantize', scheme = 'warmgreys'),
                      title='Region name'),
    detail = alt.Detail('region:N'),
    tooltip = [
        alt.Tooltip('region:O', title='Region name')
    ]
).properties(width = 600, height = 400)
```

```python
alt.hconcat(country_map, 
            side_line_plot
           ).configure_title(fontSize=17)
```
